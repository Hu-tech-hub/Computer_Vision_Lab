## 11.2 Í≥†Ïú≥Í∞íÍ≥º Í≥†Ïú†Î≤°ÌÑ∞ Í≥ÑÏÇ∞

### Í≥†Ïú†Í∞íÍ≥º Í≥†Ïú†Î≤°ÌÑ∞Ïùò Ï†ïÏùò
  * Í≥†Ïú†Í∞í(Eigenvalue)
  Ï£ºÏñ¥ÏßÑ Ï†ïÎ∞©ÌñâÎ†¨ $A$Ïóê ÎåÄÌï¥, Î≤°ÌÑ∞ $\mathbf{v}$Í∞Ä ÌñâÎ†¨ $A$Ïóê ÏùòÌï¥ Ïä§ÏπºÎùº $\lambda$Î∞∞ÎßåÌÅº Î≥ÄÌïòÎäî Í≤ΩÏö∞, Í∑∏ Ïä§ÏπºÎùº Í∞íÏùÑ ÌñâÎ†¨ $A$Ïùò Í≥†Ïú†Í∞íÏù¥ÎùºÍ≥† Ìï©ÎãàÎã§.

  * Í≥†Ïú†Î≤°ÌÑ∞(Eigenvector)
  Í≥†Ïú†Í∞í $\lambda$Ïóê ÎåÄÏùëÌïòÎäî Î≤°ÌÑ∞ $\mathbf{v}$Î•º ÌñâÎ†¨$A$Ïùò Í≥†Ïú†Î≤°ÌÑ∞ÎùºÍ≥† Ìï©ÎãàÎã§. (Í≥†Ïú†Î≤°ÌÑ∞ $\mathbf{v}$Îäî ÏòÅÎ≤°ÌÑ∞Í∞Ä ÏïÑÎãàÎã§.)

  * Í≥†Ïú†Í∞í $\lambda$ÏôÄ Í≥†Ïú†Î≤°ÌÑ∞ $\mathbf{v}$Îäî Îã§Ïùå Ï°∞Í±¥ÏùÑ ÎßåÏ°±Ìï©ÎãàÎã§.

   $$A \mathbf{v} = \lambda \mathbf{v}$$
  
  * Ïù¥ ÏãùÏùÄ ÌñâÎ†¨
  ùê¥Î•º Í≥±Ìïú Í≤∞Í≥ºÍ∞Ä Îã®ÏàúÌûà Ïä§ÏπºÎùº ùúÜÎ•º Í≥±Ìïú Î≤°ÌÑ∞ ùë£ÏôÄ Í∞ôÎã§Îäî Í≤ÉÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§. Ï¶â, ÌñâÎ†¨ ùê¥Í∞Ä Í≥†Ïú†Î≤°ÌÑ∞ ùë£Ïùò Î∞©Ìñ•ÏùÑ Î∞îÍæ∏ÏßÄ ÏïäÏúºÎ©∞, ÌÅ¨Í∏∞Îßå ùúÜÎ∞∞Î°ú Î∞îÎÄùÎãàÎã§.

---

### Í≥†Ïú†Í∞íÍ≥º Í≥†Ïú†Î≤°ÌÑ∞Ïùò Í≥ÑÏÇ∞

  * ÏúÑ Ï°∞Í±¥ÏãùÏóêÏÑú Ïö∞Î≥ÄÏóê ÏúÑÏπòÌïú Ìï≠ÏùÑ Ï¢åÎ≥ÄÏúºÎ°ú ÏòÆÍ∏∞Î©¥ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§.

  $$ (A - \lambda I) \mathbf{v} = 0 $$

  * ÏúÑ ÏãùÏóêÏÑú Í≥†Ïú≥Í∞í $\lambda$Í∞Ä Ï°¥Ïû¨ÌïòÍ∏∞ ÏúÑÌïú ÌïÑÏöîÏ∂©Î∂ÑÏ°∞Í±¥ÏùÄ $(A - \lambda I)$Ïùò ÌñâÎ†¨ÏãùÏù¥ 0Ïù¥ ÎêòÎäî Í≤ÉÏûÖÎãàÎã§. ($I$Îäî Îã®ÏúÑÎ≤°ÌÑ∞ÏûÖÎãàÎã§.)
  

  $$\det(A - \lambda I) = 0$$

  * ÏúÑÏôÄ Í∞ôÏùÄ ÏãùÏùÑ ÌäπÏÑ±Î∞©Ï†ïÏãùÏù¥ÎùºÍ≥† ÌïòÎ©∞, ÌäπÏÑ± Î∞©Ï†ïÏãùÏùÑ ÌíÄÎ©¥ Í≥†Ïú†Í∞í $\lambda$Î•º Íµ¨Ìï† Ïàò ÏûàÏäµÎãàÎã§.

---

### ÏòàÏ†ú

  #### 1. Í≥†Ïú†Í∞í $\lambda$ Íµ¨ÌïòÍ∏∞

  Ï†ïÎ∞© ÌñâÎ†¨ $A$Í∞Ä Îã§ÏùåÍ≥º Í∞ôÎã§Í≥† Í∞ÄÏ†ïÌï¥Î¥ÖÏãúÎã§.

  $$
  A = \begin{bmatrix}
  4 & 1 \\
  2 & 3
  \end{bmatrix}
  $$

  ÌäπÏÑ± Î∞©Ï†ïÏãù $\det(A - \lambda I) = 0$ÏùÑ ÌíÄÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§.

  $$
  \det \begin{bmatrix}
  4 - \lambda & 1 \\
  2 & 3 - \lambda
  \end{bmatrix} = 0
  $$

  ÌñâÎ†¨ÏãùÏùÑ Í≥ÑÏÇ∞ÌïòÎ©¥,

  $$
  (4 - \lambda)(3 - \lambda) - 2 \cdot 1 = 0
  $$

  $$
  \lambda^2 - 7\lambda + 10 = 0
  $$

  Ïù¥ Î∞©Ï†ïÏãùÏùÑ ÌíÄÎ©¥,

  $$
  \lambda = 5, \quad \lambda = 2
  $$

  ---

  #### 2. Í≥†Ïú†Í∞í $\lambda = 2$Ïóê ÎåÄÏùëÌïòÎäî Í≥†Ïú†Î≤°ÌÑ∞Î•º Íµ¨ÌïòÍ∏∞.

  Í≥†Ïú†Î≤°ÌÑ∞Î•º Ï∞æÎäî Í≥ºÏ†ïÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùÄ ÎèôÏ∞® ÏÑ†Ìòï ÏãúÏä§ÌÖúÏùÑ ÎßåÏ°±ÌïòÎäî $\mathbf{v}$Î•º Ï∞æÎäî Í≤ÉÏûÖÎãàÎã§.

  $$
  (A - 2I) \mathbf{v} = 0
  $$

  $$
  \begin{bmatrix}
  4 - 2 & 1 \\
  2 & 3 - 2
  \end{bmatrix}
  \mathbf{v} = \begin{bmatrix}
  2 & 1 \\
  2 & 1
  \end{bmatrix}
  \begin{bmatrix}
  v_1 \\
  v_2
  \end{bmatrix} = \begin{bmatrix}
  0 \\
  0
  \end{bmatrix}
  $$

  Ïù¥ ÌñâÎ†¨ÏùÑ ÌíÄÎ©¥,

  $$
  2v_1 + v_2 = 0 \implies v_2 = -2v_1
  $$

  Îî∞ÎùºÏÑú, $\lambda = 2$ Ïóê ÎåÄÏùëÌïòÎäî Í≥†Ïú†Î≤°ÌÑ∞ $\mathbf{v}$ Îäî Îã§ÏùåÍ≥º Í∞ôÏäµÎãàÎã§.

  
  $$\mathbf{v} = k \begin{bmatrix} 1 \\ -2 \end{bmatrix}
  (Ïó¨Í∏∞ÏÑú kÎäî ÏûÑÏùòÏùò ÏÉÅÏàò)$$

  ---

  #### 3. Í≥†Ïú†Í∞í $\lambda = 5$Ïóê ÎåÄÏùëÌïòÎäî Í≥†Ïú†Î≤°ÌÑ∞Î•º Íµ¨ÌïòÍ∏∞

  $$
  (A - 5I) \mathbf{v} = 0
  $$

  $$
  \begin{bmatrix}
  4 - 5 & 1 \\
  2 & 3 - 5
  \end{bmatrix}
  \mathbf{v} = \begin{bmatrix}
  -1 & 1 \\
  2 & -2
  \end{bmatrix}
  \begin{bmatrix}
  v_1 \\
  v_2
  \end{bmatrix} = \begin{bmatrix}
  0 \\
  0
  \end{bmatrix}
  $$

  Ïù¥ ÌñâÎ†¨ÏùÑ ÌíÄÎ©¥,

  $$
  v_1 + v_2 = 0 \implies v_2 = v_1
  $$

  Îî∞ÎùºÏÑú, $\lambda = 5$Ïóê ÎåÄÏùëÌïòÎäî Í≥†Ïú†Î≤°ÌÑ∞ $\mathbf{v}$Îäî Îã§ÏùåÍ≥º Í∞ôÎã§.
  
  $$\mathbf{v} = k \begin{bmatrix} 1 \\ 1 \end{bmatrix}
  (Ïó¨Í∏∞ÏÑú kÎäî ÏûÑÏùòÏùò ÏÉÅÏàò)$$



```python
import numpy as np
import matplotlib.pyplot as plt

# ÌñâÎ†¨ A Ï†ïÏùò
A = np.array([[2, 1],
              [1, 2]])

# Í≥†Ïú†Í∞í Î∞è Í≥†Ïú†Î≤°ÌÑ∞ Í≥ÑÏÇ∞
eigenvalues, eigenvectors = np.linalg.eig(A)

# ÏãúÍ∞ÅÌôî
fig, ax = plt.subplots()

# ÏõêÏ†êÏùÑ Ï§ëÏã¨ÏúºÎ°ú Ìïú Î≤°ÌÑ∞Îì§
origin = np.zeros((2, eigenvectors.shape[1]))  # (2, 2) ÌÅ¨Í∏∞Ïùò Î∞∞Ïó¥

# ÌñâÎ†¨Ïùò Í≥†Ïú†Î≤°ÌÑ∞
ax.quiver(*origin, eigenvectors[0, :], eigenvectors[1, :], color=['r', 'b'], angles='xy', scale_units='xy', scale=1, label='Eigenvectors')

# Î≥ÄÌôòÎêú Í≥†Ïú†Î≤°ÌÑ∞
transformed_vectors = A @ eigenvectors
ax.quiver(*origin, transformed_vectors[0, :], transformed_vectors[1, :], color=['r', 'b'], alpha=0.5, angles='xy', scale_units='xy', scale=1, label='Transformed Eigenvectors')

# ÌñâÎ†¨ AÏùò Í≥†Ïú†Í∞íÍ≥º Í≥†Ïú†Î≤°ÌÑ∞ ÌëúÏãú
for i in range(len(eigenvalues)):
    ax.text(eigenvectors[0, i]*1.1, eigenvectors[1, i]*1.1, f'Œª={eigenvalues[i]:.2f}', fontsize=12, color='k')
    ax.text(transformed_vectors[0, i]*1.1, transformed_vectors[1, i]*1.1, f'Œªv', fontsize=12, color='gray')

# Ï∂ïÏùò Î≤îÏúÑ ÏÑ§Ï†ï
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

# Ï∂ï Î†àÏù¥Î∏î ÏÑ§Ï†ï
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.grid()
ax.set_aspect('equal')

# Î≤îÎ°Ä Ï∂îÍ∞Ä
ax.legend()

plt.title('Eigenvalues and Eigenvectors Transformation')
plt.show()
```
