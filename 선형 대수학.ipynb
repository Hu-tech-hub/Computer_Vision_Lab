# 11. 고윳값과 고유 벡터의 개념

## 11.2 고윳값과 고유벡터 계산

### 고유값과 고유벡터의 정의
  * 고유값(Eigenvalue)
  주어진 정방행렬 $A$에 대해, 벡터 $\mathbf{v}$가 행렬 $A$에 의해 스칼라 $\lambda$배만큼 변하는 경우, 그 스칼라 값을 행렬 $A$의 고유값이라고 합니다.

  * 고유벡터(Eigenvector)
  고유값 $\lambda$에 대응하는 벡터 $\mathbf{v}$를 행렬$A$의 고유벡터라고 합니다. (고유벡터 $\mathbf{v}$는 영벡터가 아니다.)

  * 고유값 $\lambda$와 고유벡터 $\mathbf{v}$는 다음 조건을 만족합니다.

   $$A \mathbf{v} = \lambda \mathbf{v}$$
  
  * 이 식은 행렬
  𝐴를 곱한 결과가 단순히 스칼라 𝜆를 곱한 벡터 𝑣와 같다는 것을 의미합니다. 즉, 행렬 𝐴가 고유벡터 𝑣의 방향을 바꾸지 않으며, 크기만 𝜆배로 바뀝니다.

---

### 고유값과 고유벡터의 계산

  * 위 조건식에서 우변에 위치한 항을 좌변으로 옮기면 다음과 같습니다.

  $$ (A - \lambda I) \mathbf{v} = 0 $$

  * 위 식에서 고윳값 $\lambda$가 존재하기 위한 필요충분조건은 $(A - \lambda I)$의 행렬식이 0이 되는 것입니다. ($I$는 단위벡터입니다.)
  

  $$\det(A - \lambda I) = 0$$

  * 위와 같은 식을 특성방정식이라고 하며, 특성 방정식을 풀면 고유값 $\lambda$를 구할 수 있습니다.

---

### 예제

  #### 1. 고유값 $\lambda$ 구하기

  정방 행렬 $A$가 다음과 같다고 가정해봅시다.

  $$
  A = \begin{bmatrix}
  4 & 1 \\
  2 & 3
  \end{bmatrix}
  $$

  특성 방정식 $\det(A - \lambda I) = 0$을 풀면 다음과 같습니다.

  $$
  \det \begin{bmatrix}
  4 - \lambda & 1 \\
  2 & 3 - \lambda
  \end{bmatrix} = 0
  $$

  행렬식을 계산하면,

  $$
  (4 - \lambda)(3 - \lambda) - 2 \cdot 1 = 0
  $$

  $$
  \lambda^2 - 7\lambda + 10 = 0
  $$

  이 방정식을 풀면,

  $$
  \lambda = 5, \quad \lambda = 2
  $$

  ---

  #### 2. 고유값 $\lambda = 2$에 대응하는 고유벡터를 구하기.

  고유벡터를 찾는 과정은 다음과 같은 동차 선형 시스템을 만족하는 $\mathbf{v}$를 찾는 것입니다.

  $$
  (A - 2I) \mathbf{v} = 0
  $$

  $$
  \begin{bmatrix}
  4 - 2 & 1 \\
  2 & 3 - 2
  \end{bmatrix}
  \mathbf{v} = \begin{bmatrix}
  2 & 1 \\
  2 & 1
  \end{bmatrix}
  \begin{bmatrix}
  v_1 \\
  v_2
  \end{bmatrix} = \begin{bmatrix}
  0 \\
  0
  \end{bmatrix}
  $$

  이 행렬을 풀면,

  $$
  2v_1 + v_2 = 0 \implies v_2 = -2v_1
  $$

  따라서, $\lambda = 2$ 에 대응하는 고유벡터 $\mathbf{v}$ 는 다음과 같습니다.

  
  $$\mathbf{v} = k \begin{bmatrix} 1 \\ -2 \end{bmatrix}
  (여기서 k는 임의의 상수)$$

  ---

  #### 3. 고유값 $\lambda = 5$에 대응하는 고유벡터를 구하기

  $$
  (A - 5I) \mathbf{v} = 0
  $$

  $$
  \begin{bmatrix}
  4 - 5 & 1 \\
  2 & 3 - 5
  \end{bmatrix}
  \mathbf{v} = \begin{bmatrix}
  -1 & 1 \\
  2 & -2
  \end{bmatrix}
  \begin{bmatrix}
  v_1 \\
  v_2
  \end{bmatrix} = \begin{bmatrix}
  0 \\
  0
  \end{bmatrix}
  $$

  이 행렬을 풀면,

  $$
  v_1 + v_2 = 0 \implies v_2 = v_1
  $$

  따라서, $\lambda = 5$에 대응하는 고유벡터 $\mathbf{v}$는 다음과 같다.
  
  $$\mathbf{v} = k \begin{bmatrix} 1 \\ 1 \end{bmatrix}
  (여기서 k는 임의의 상수)$$



```python
import numpy as np
import matplotlib.pyplot as plt

# 행렬 A 정의
A = np.array([[2, 1],
              [1, 2]])

# 고유값 및 고유벡터 계산
eigenvalues, eigenvectors = np.linalg.eig(A)

# 시각화
fig, ax = plt.subplots()

# 원점을 중심으로 한 벡터들
origin = np.zeros((2, eigenvectors.shape[1]))  # (2, 2) 크기의 배열

# 행렬의 고유벡터
ax.quiver(*origin, eigenvectors[0, :], eigenvectors[1, :], color=['r', 'b'], angles='xy', scale_units='xy', scale=1, label='Eigenvectors')

# 변환된 고유벡터
transformed_vectors = A @ eigenvectors
ax.quiver(*origin, transformed_vectors[0, :], transformed_vectors[1, :], color=['r', 'b'], alpha=0.5, angles='xy', scale_units='xy', scale=1, label='Transformed Eigenvectors')

# 행렬 A의 고유값과 고유벡터 표시
for i in range(len(eigenvalues)):
    ax.text(eigenvectors[0, i]*1.1, eigenvectors[1, i]*1.1, f'λ={eigenvalues[i]:.2f}', fontsize=12, color='k')
    ax.text(transformed_vectors[0, i]*1.1, transformed_vectors[1, i]*1.1, f'λv', fontsize=12, color='gray')

# 축의 범위 설정
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

# 축 레이블 설정
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.grid()
ax.set_aspect('equal')

# 범례 추가
ax.legend()

plt.title('Eigenvalues and Eigenvectors Transformation')
plt.show()
```
